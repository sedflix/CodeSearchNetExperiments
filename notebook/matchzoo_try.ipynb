```python
all_langs = ["python", "java", "go", "php", "ruby", "javascript"]
```

```python
df = get_data_df("../resources/data/", all_langs[0:1] , ["train", "valid"])
print(len(df))
```

```python
def get_data_pack(df_):
    till = len(df_)
    querys = df_.docstring_tokens.apply(lambda x : " ".join(x)).to_list()
    codes = df_.code_tokens.apply(lambda x : " ".join(x)).to_list()
    labels = [1 for _ in range(till)]
    for i in range(till // 2):
        querys.append(querys[random.randint(0, till)]) 
        codes.append(codes[random.randint(0, till)]) 
        labels.append(0)
    data = pd.DataFrame({
    'text_left': querys, 
    'text_right':codes, 
    'label': labels})
    return mz.pack(data)
```

```python
train_pack = get_data_pack(df[df['partition']=='train'])
val_pack = get_data_pack(df[df['partition']=='valid'])
```

```python
preprocessor = mz.preprocessors.BasicPreprocessor(
    fixed_length_left=28,
    fixed_length_right = 268,
    filter_low_freq = 10,
    filter_mode='tf'
)
preprocessor.fit(train_pack)
train_pack = preprocessor.transform(train_pack)
val_pack = preprocessor.transform(val_pack)
```

```python
train_pack.frame[0]
```

```python
ranking_task = mz.tasks.Ranking(loss=mz.losses.RankHingeLoss())
ranking_task.metrics = [
    mz.metrics.NormalizedDiscountedCumulativeGain(k=3),
    mz.metrics.NormalizedDiscountedCumulativeGain(k=1000),
    mz.metrics.MeanReciprocalRank()
]
print("`ranking_task` initialized with metrics", ranking_task.metrics)
```

```python
preprocessor.context
```

```python
model = mz.models.KNRM()
model.params.update(preprocessor.context)
model.params['task'] = ranking_task
model.params['embedding_output_dim'] = 200
model.params['embedding_trainable'] = True
model.params['kernel_num'] = 21
model.params['sigma'] = 0.1
model.params['exact_sigma'] = 0.001
model.params['optimizer'] = 'adadelta'
model.build()
model.compile()
```
